```python
class vehicle:
    company="tayota"
    brand="Inova"
 
    def vehicle_brand(self):
        print('company')
    def vehicle_company(self):
        print('brand')
obj=vehicle()
obj.vehicle_brand()
obj.vehicle_company()
```

```python
class operations:
    "Arthematic operations"
    a=10
    b=20
    def add(self):
        print("Addition of %d+%d=%d"%(self.a,self.b,self.a+self.b))
    def sub(self):
        print("Substraction of %d-%d=%d"%(self.a,self.b,self.a-self.b))
obj=operations()
obj.add()
obj.sub()
 
```
